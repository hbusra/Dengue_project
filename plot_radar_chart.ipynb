```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams


def plot_radar_expression(df1, df2, groups=['adults', 'children'], figsize=(6,6), rcircle=3.2,
                          geneset=None, ct='NK cells', title='Radar Plot', name=''):
    if geneset is None:
        geneset = [['genes']]
        
    rcParams.update({
        'font.size': 16,
        'pdf.fonttype': 42,
        'font.sans-serif': 'Arial',
        'figure.dpi': 100
    })

    colors = [(0.91, 0.54, 0.76), (0.99, 0.55, 0.38)]
    fig = plt.figure(figsize=figsize)

    for i, genes in enumerate(geneset, start=1):
        categories = [*genes, genes[0]]
        values1 = df1.loc[ct].loc[categories]
        values2 = df2.loc[ct].loc[categories]
        
        angles = np.linspace(start=0, stop=2 * np.pi, num=len(categories))
        
        ax = fig.add_subplot(1, len(geneset), i, polar=True)
        ax.plot(angles, values1, label=groups[0], color=colors[0], linewidth=2)
        ax.plot(angles, values2, label=groups[1], color=colors[1], linewidth=2)

        ax.set_thetagrids(np.degrees(angles[:-1]), labels=categories[:-1])
        ax.set_rlabel_position(-15)
        ax.set_rticks([-3, -2, -1, 0, 1, 2, 3])
        ax.grid(True, ls='--', lw=0.5)

        ax.add_artist(plt.Circle((0, 0), rcircle, transform=ax.transData._b, edgecolor="k",
                                 fill=False, lw=1))

        ax.set_title(name.replace('_', ' '), size=14, y=1.1)

    fig.suptitle(title, fontsize=16)
    fig.text(0.3, -0.1, f'log2 fold change (SDp vs D)\n          {ct}', fontsize=16)
    fig.tight_layout()
    fig.legend(groups, bbox_to_anchor=(1.1, 1.05), loc='upper right')
    
    return {'fig': fig, 'ax': ax}
```
